```python
import pandas as pd

data = {
    "Product": ["Milk", "Bread", "Rice", "Eggs", "Sugar"],
    "Sales": [100, 150, 200, 120, 80],
    "Branch": ["A", "B", "A", "C", "B"]
}

df = pd.DataFrame(data)

df
```

```python
df["Sales"].sum()
```

```python
df.groupby("Branch")["Sales"].sum()
```

```python
import matplotlib.pyplot as plt

df.groupby("Branch")["Sales"].sum().plot(kind="bar")

plt.title("Sales by Branch")
plt.show()
```

```python
import pandas as pd

data = {
    "Product": [
        "Milk","Bread","Rice","Eggs","Sugar",
        "Oil","Soap","Biscuits","Tea","Coffee",
        "Milk","Bread","Rice","Eggs","Sugar",
        "Oil","Soap","Biscuits","Tea","Coffee"
    ],
    "Sales": [
        100,150,200,120,80,
        180,90,140,160,220,
        130,170,250,110,95,
        190,85,145,175,210
    ],
    "Branch": [
        "A","B","A","C","B",
        "A","C","B","A","C",
        "B","A","C","A","B",
        "C","B","A","C","A"
    ],
    "Payment": [
        "Cash","Card","UPI","Cash","Card",
        "UPI","Cash","Card","UPI","Cash",
        "Card","UPI","Cash","Card","UPI",
        "Cash","Card","UPI","Cash","Card"
    ],
    "Rating": [
        8.5,9.0,8.8,7.9,8.2,
        9.1,7.8,8.4,8.9,9.2,
        8.6,9.1,8.7,8.0,8.3,
        9.0,7.7,8.5,8.8,9.3
    ]
}

df = pd.DataFrame(data)

df.head()
```

```python
df["Sales"].sum()
```

```python
df.groupby("Branch")["Sales"].sum()
```

```python
df.groupby("Product")["Sales"].sum().sort_values(ascending=False)
```

```python
df["Payment"].value_counts()
```

```python
df.groupby("Branch")["Rating"].mean()
```

```python
import matplotlib.pyplot as plt

df.groupby("Branch")["Sales"].sum().plot(kind="bar")
plt.title("Sales by Branch")
plt.show()
```

```python
df["Payment"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.ylabel("")
plt.title("Payment Methods")
plt.show()
```

```python
df.groupby("Product")["Sales"].sum().sort_values().plot(kind="barh")
plt.title("Sales by Product")
plt.show()
```

```python
df.groupby("Branch")["Rating"].mean().plot(kind="bar")
plt.title("Average Rating by Branch")
plt.show()
```
